**Ejercicio 1 — Calentamiento**  
Dado este diccionario, imprime cada gen y su longitud:
`genes = {     
"TP53": "ATGGATGGTGCAGAGTTCAAACGGCAGACTGGTCCTGAGCAGGGCCAGGA",    
"EGFR": "ATGCGACCCTCCGGGACGGCCGGGGCAGCGCTCCTGGCGCTGCTGGCTGC",    
"ALK":  "ATGATTCAAGAAAGCAAAGCAATGGCGGCGGCAGCAGCACCAGCAGCAGCA" }`

```python
genes = {
"TP53": "ATGGATGGTGCAGAGTTCAAACGGCAGACTGGTCCTGAGCAGGGCCAGGA",
"EGFR": "ATGCGACCCTCCGGGACGGCCGGGGCAGCGCTCCTGGCGCTGCTGGCTGC",
"ALK":  "ATGATTCAAGAAAGCAAAGCAATGGCGGCGGCAGCAGCACCAGCAGCAGCA" 
}
```

```python
#Leer diccionario
for gen,secuencia in genes.items():
    print(f'El gen {gen} tiene una longitud de {(len(secuencia))} nucleótidos')
```

**Ejercicio 2 — Función**  
Escribe una función `contar_adenina(secuencia)` que devuelva el porcentaje de A en una secuencia. Aplícala a cada gen del diccionario anterior.

```python
#Definir la funcion
def contar_adenina(secuencia):
    porcentaje_de_adenina = (secuencia.count('A')/len(secuencia))*100
    return porcentaje_de_adenina
#Leer diccionariu
for gen,secuencia in genes.items():
#Aplicar la nueva función y guardarla en una variable
    porcentaje = contar_adenina(secuencia)
    print(f'El gen {gen} tiene un porcentaje de adenina del {porcentaje:.1f}%')
```

**Ejercicio 3 — Comparator**  
Usando el mismo diccionario, encuentra el gen con **mayor porcentaje de adenina**. Contempla empates.

```python
#Creamos una variable como comparador
comparator = 0
#Leer diccionario
for gen,secuencia in genes.items():
#Comparar los porcentajes y quedarnos con el más alto. Si empatan 2 con el valor más alto se guardan.
    if porcentaje>comparator:
        comparator = porcentaje
        max_gen = gen
    elif secuencia == comparator:
        max_gen.append(gen)
#Imprimirlo
print(f'El gen con mayor % de adenina es {max_gen} con un total del {comparator}%')
```

**Ejercicio 4 — Lista por comprensión**  
Crea una lista con los nombres de los genes que tengan **más de 10 adeninas** en su secuencia.

```python
genes_10_adeninas=[gen for gen,secuencia in genes.items()if secuencia.count('A')>10]
print(genes_10_adeninas)
```

**Ejercicio 5 — Reto integrador**  
Crea un **nuevo diccionario** donde las claves sean los nombres de los genes y los valores sean su %GC redondeado a 2 decimales. Pista: es como una lista por comprensión pero con `{}` y `clave: valor`.

```python
nuevo_dic = {gen: round((secuencia.count('G') + secuencia.count('C')) / len(secuencia) * 100, 2)
           for gen, secuencia in genes.items()}
print(nuevo_dic)
```
